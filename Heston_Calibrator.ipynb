```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tqdm import tqdm
from scipy.stats import norm
from scipy.optimize import newton
import multiprocessing as mp
from functools import partial
import time
import os
import gc
from mpl_toolkits.mplot3d import Axes3D

# --- 1. GLOBAL CONFIGURATION ---
S0, v0, r0 = 100.0, 0.04, 0.05
kappa_r, theta_r, xi_r = 0.5, 0.05, 0.01 
TRAIN_STRIKES = np.array([85, 90, 95, 100, 105, 110, 115], dtype=np.float32)
MATURITIES = [0.1, 0.5, 1.0, 2.0, 3.0]
dt = 1/252

# Memory-safe core allocation
N_CORES = max(1, min(8, mp.cpu_count() - 2))  # Cap at 8 cores
print(f"Using {N_CORES} cores (system has {mp.cpu_count()})")

# --- 2. ANALYTIC UTILITIES ---
def get_bs_price(S, K, T, r, sigma):
    if T <= 1e-6: return max(S - K, 0)
    d1 = (np.log(S / K) + (r + 0.5 * sigma**2) * T) / (sigma * np.sqrt(T))
    d2 = d1 - sigma * np.sqrt(T)
    return S * norm.cdf(d1) - K * np.exp(-r * T) * norm.cdf(d2)

def find_iv(target_price, S, K, T, r):
    intrinsic = max(S - K * np.exp(-r * T), 0)
    if target_price <= intrinsic + 1e-7: return 0.001
    try:
        return newton(lambda x: get_bs_price(S, K, T, r, x) - target_price, 0.2, tol=1e-5, maxiter=50)
    except: 
        return np.nan

def get_vega(S, K, T, r, sigma):
    if T <= 0 or sigma <= 0: return 0
    d1 = (np.log(S / K) + (r + 0.5 * sigma**2) * T) / (sigma * np.sqrt(T))
    return S * np.sqrt(T) * norm.pdf(d1)

# --- 3. MEMORY-EFFICIENT SDE ENGINE ---
def simulate_heston_hw(params, n_paths=5000, strikes=TRAIN_STRIKES, seed=None):
    """Memory-optimized Heston simulator"""
    if seed is not None:
        np.random.seed(seed)
    
    k_v, t_v, x_v, rho = params
    n_steps = int(max(MATURITIES) / dt)
    z_corr = np.array([[1., rho, 0.], [rho, 1., 0.], [0., 0., 1.]], dtype=np.float32)
    L = np.linalg.cholesky(z_corr)
    
    # Use float32 to halve memory usage
    S = np.full(n_paths, S0, dtype=np.float32)
    v = np.full(n_paths, v0, dtype=np.float32)
    r = np.full(n_paths, r0, dtype=np.float32)
    prices = []
    mat_steps = {int(round(T / dt)): T for T in MATURITIES}
    
    for step in range(1, n_steps + 1):
        W = (L @ np.random.standard_normal((3, n_paths))).astype(np.float32)
        vp = np.maximum(v, 1e-6)
        r += kappa_r*(theta_r-r)*dt + xi_r*np.sqrt(dt)*W[2]
        S *= np.exp((r - 0.5*vp)*dt + np.sqrt(vp*dt)*W[0])
        v += k_v*(t_v-vp)*dt + x_v*np.sqrt(vp*dt)*W[1]
        
        if step in mat_steps:
            T = mat_steps[step]
            payoff = np.maximum(S[None, :] - strikes[:, None], 0)
            prices.extend((payoff * np.exp(-r*T)).mean(axis=1) / S0)
    
    return np.array(prices, dtype=np.float32)

# --- 4. CHUNKED DATA GENERATION ---
def generate_sample_worker(sample_id):
    """Generate single sample (memory-efficient)"""
    np.random.seed(sample_id)
    
    for _ in range(100):
        k, t, x, r = np.random.uniform(0.5, 4.0), np.random.uniform(0.01, 0.1), \
                     np.random.uniform(0.1, 0.6), np.random.uniform(-0.9, -0.2)
        if (2*k*t) > x**2: break
    else:
        k, t, x, r = 1.5, 0.04, 0.2, -0.5
    
    # Reduce paths for training data to save memory
    prices = simulate_heston_hw([k, t, x, r], n_paths=3000, seed=sample_id + 100000)
    return prices, np.array([k, t, x, r], dtype=np.float32)

def generate_data_in_chunks(n_samples, chunk_size=500, cache_dir='./data_cache'):
    """
    Generate data in chunks and cache to disk
    
    Args:
        n_samples: Total samples to generate
        chunk_size: Samples per chunk (smaller = less RAM)
        cache_dir: Directory for temporary cache files
    """
    os.makedirs(cache_dir, exist_ok=True)
    n_chunks = int(np.ceil(n_samples / chunk_size))
    
    print(f"\nGenerating {n_samples} samples in {n_chunks} chunks of {chunk_size}...")
    
    for chunk_idx in range(n_chunks):
        start_idx = chunk_idx * chunk_size
        end_idx = min(start_idx + chunk_size, n_samples)
        chunk_samples = end_idx - start_idx
        
        print(f"\nChunk {chunk_idx + 1}/{n_chunks} ({chunk_samples} samples)")
        
        # Generate chunk in parallel
        with mp.Pool(N_CORES) as pool:
            results = list(tqdm(
                pool.imap(generate_sample_worker, range(start_idx, end_idx)),
                total=chunk_samples,
                desc=f"Chunk {chunk_idx + 1}"
            ))
        
        # Separate and save to disk
        X_chunk, Y_chunk = zip(*results)
        X_chunk = np.array(X_chunk, dtype=np.float32)
        Y_chunk = np.array(Y_chunk, dtype=np.float32)
        
        np.savez_compressed(
            f'{cache_dir}/chunk_{chunk_idx}.npz',
            X=X_chunk,
            Y=Y_chunk
        )
        
        # Free memory immediately
        del X_chunk, Y_chunk, results
        gc.collect()
    
    print(f"✓ All chunks cached to {cache_dir}/")

def load_data_from_cache(cache_dir='./data_cache'):
    """Load all cached chunks and combine"""
    chunk_files = sorted([f for f in os.listdir(cache_dir) if f.startswith('chunk_')])
    
    X_list, Y_list = [], []
    print(f"\nLoading {len(chunk_files)} chunks from cache...")
    
    for chunk_file in tqdm(chunk_files, desc="Loading"):
        data = np.load(f'{cache_dir}/{chunk_file}')
        X_list.append(data['X'])
        Y_list.append(data['Y'])
    
    X_data = np.vstack(X_list)
    Y_data = np.vstack(Y_list)
    
    print(f"✓ Loaded {len(X_data)} samples")
    return X_data, Y_data

# --- 5. MAIN EXECUTION ---
if __name__ == "__main__":
    N_SAMPLES = 10000
    CHUNK_SIZE = 500  # Process 500 samples at a time
    CACHE_DIR = './heston_cache'
    
    print(f"\n{'='*60}")
    print(f"MEMORY-EFFICIENT PARALLEL TRAINING")
    print(f"{'='*60}")
    print(f"Total samples: {N_SAMPLES}")
    print(f"Chunk size: {CHUNK_SIZE}")
    print(f"Cores: {N_CORES}")
    print(f"Cache directory: {CACHE_DIR}")
    
    # Generate or load data
    if os.path.exists(CACHE_DIR) and len(os.listdir(CACHE_DIR)) > 0:
        print(f"\n⚠ Found existing cache. Delete '{CACHE_DIR}' to regenerate.")
        response = input("Use cached data? (y/n): ").strip().lower()
        if response == 'y':
            X_data, Y_data = load_data_from_cache(CACHE_DIR)
        else:
            print("Regenerating data...")
            import shutil
            shutil.rmtree(CACHE_DIR)
            generate_data_in_chunks(N_SAMPLES, CHUNK_SIZE, CACHE_DIR)
            X_data, Y_data = load_data_from_cache(CACHE_DIR)
    else:
        generate_data_in_chunks(N_SAMPLES, CHUNK_SIZE, CACHE_DIR)
        X_data, Y_data = load_data_from_cache(CACHE_DIR)
    
    # Data quality check
    print(f"\nData Quality:")
    print(f"  Shape: X={X_data.shape}, Y={Y_data.shape}")
    print(f"  Memory: X={X_data.nbytes/1e6:.1f}MB, Y={Y_data.nbytes/1e6:.1f}MB")
    
    # Scaling
    X_m, X_s = X_data.mean(0).astype(np.float32), X_data.std(0).astype(np.float32)
    Y_m, Y_s = Y_data.mean(0).astype(np.float32), Y_data.std(0).astype(np.float32)
    
    # Normalize in-place to save memory
    X_data = (X_data - X_m) / X_s
    Y_data = (Y_data - Y_m) / Y_s
    
    # --- 6. NEURAL NETWORK ---
    print(f"\n{'='*60}")
    print(f"TRAINING NEURAL CALIBRATOR")
    print(f"{'='*60}")
    
    model = tf.keras.Sequential([
        tf.keras.layers.Input(shape=(X_data.shape[1],)),
        tf.keras.layers.Dense(128, activation='elu'),  # Smaller network
        tf.keras.layers.BatchNormalization(),
        tf.keras.layers.Dropout(0.1),
        tf.keras.layers.Dense(64, activation='elu'),
        tf.keras.layers.Dense(4)
    ])
    
    model.compile(
        optimizer=tf.keras.optimizers.Adam(0.001),
        loss='mse',
        metrics=['mae']
    )
    
    # Train with smaller batch size
    callbacks = [
        tf.keras.callbacks.EarlyStopping(patience=15, restore_best_weights=True),
        tf.keras.callbacks.ReduceLROnPlateau(factor=0.5, patience=8)
    ]
    
    history = model.fit(
        X_data, Y_data,
        epochs=100,
        batch_size=32,  # Smaller batches
        validation_split=0.15,
        callbacks=callbacks,
        verbose=1
    )
    
    # --- 7. VALIDATION (MEMORY-SAFE) ---
    print(f"\n{'='*60}")
    print(f"VALIDATION")
    print(f"{'='*60}")
    
    # Only predict on subset to save memory
    test_size = 500
    P_raw = model.predict(X_data[-test_size:], verbose=0, batch_size=32)
    P = (P_raw * Y_s) + Y_m
    
    # Denormalize test targets
    T_test = (Y_data[-test_size:] * Y_s) + Y_m
    
    # Apply constraints
    P[:, 0] = np.clip(P[:, 0], 0.3, 5.0)
    P[:, 1] = np.clip(P[:, 1], 0.005, 0.15)
    P[:, 2] = np.clip(P[:, 2], 0.05, 0.8)
    P[:, 3] = np.clip(P[:, 3], -0.99, -0.05)
    
    # Metrics
    param_names = ['κ_v', 'θ_v', 'ξ_v', 'ρ']
    print("\nParameter Accuracy:")
    for i, name in enumerate(param_names):
        mae = np.mean(np.abs(T_test[:, i] - P[:, i]))
        print(f"  {name}: MAE={mae:.4f}")
    
    # --- 8. VISUALIZATION ---
    fig, axes = plt.subplots(2, 3, figsize=(15, 10))
    
    # Loss
    ax = axes[0, 0]
    ax.plot(history.history['loss'], label='Train')
    ax.plot(history.history['val_loss'], label='Val')
    ax.set_yscale('log')
    ax.set_title('Training Loss')
    ax.legend()
    ax.grid(True, alpha=0.3)
    
    # Parameter scatter plots
    for i in range(4):
        ax = axes.flatten()[i + 1]
        ax.scatter(T_test[:, i], P[:, i], s=3, alpha=0.4, color='teal')
        ax.plot([T_test[:, i].min(), T_test[:, i].max()],
                [T_test[:, i].min(), T_test[:, i].max()], 'r--')
        ax.set_title(f'{param_names[i]} Calibration')
        ax.set_xlabel('True')
        ax.set_ylabel('Predicted')
        ax.grid(True, alpha=0.3)
    
    # MAE summary
    ax = axes[1, 2]
    maes = [np.mean(np.abs(T_test[:, i] - P[:, i])) for i in range(4)]
    ax.bar(param_names, maes, color='steelblue', alpha=0.7)
    ax.set_title('Parameter MAE')
    ax.set_ylabel('MAE')
    ax.grid(True, alpha=0.3, axis='y')
    
    plt.tight_layout()
    plt.savefig('calibration_results.png', dpi=120)
    plt.show()
    
    # --- 9. SINGLE VALIDATION SAMPLE ---
    print(f"\n{'='*60}")
    print(f"PRICING VALIDATION")
    print(f"{'='*60}")
    
    idx = np.random.randint(0, test_size)
    true_p, pred_p = T_test[idx], P[idx]
    
    print(f"Sample {idx}:")
    print(f"  True: κ={true_p[0]:.3f}, θ={true_p[1]:.4f}, ξ={true_p[2]:.3f}, ρ={true_p[3]:.3f}")
    print(f"  Pred: κ={pred_p[0]:.3f}, θ={pred_p[1]:.4f}, ξ={pred_p[2]:.3f}, ρ={pred_p[3]:.3f}")
    
    # Price with reduced paths to save memory
    K_val = np.linspace(85, 115, 8)  # Fewer strikes
    
    print("\nPricing surfaces (20K paths)...")
    tp_prices = simulate_heston_hw(true_p, n_paths=20000, strikes=K_val, seed=42)
    pred_prices = simulate_heston_hw(pred_p, n_paths=20000, strikes=K_val, seed=42)
    
    tp_prices = tp_prices.reshape(len(MATURITIES), -1)
    pred_prices = pred_prices.reshape(len(MATURITIES), -1)
    
    # Compute IV surface
    iv_surf = np.array([[find_iv(p*S0, S0, k, t, r0) for p, k in zip(row, K_val)]
                        for row, t in zip(tp_prices, MATURITIES)])
    
    # Plot
    fig = plt.figure(figsize=(18, 6))
    
    # 3D IV Surface
    ax1 = fig.add_subplot(131, projection='3d')
    K_mesh, T_mesh = np.meshgrid(K_val, MATURITIES)
    surf = ax1.plot_surface(K_mesh, T_mesh, iv_surf, cmap='viridis', alpha=0.8)
    ax1.set_title("Volatility Surface")
    ax1.set_xlabel("Strike")
    ax1.set_ylabel("Maturity")
    fig.colorbar(surf, ax=ax1, shrink=0.6)
    
    # Price comparison
    ax2 = fig.add_subplot(132)
    ax2.plot(tp_prices.flatten(), 'k--', label='True', linewidth=2)
    ax2.plot(pred_prices.flatten(), 'ro', label='Predicted', markersize=4)
    ax2.set_title("Price Replication")
    ax2.set_xlabel("Option Index")
    ax2.set_ylabel("Normalized Price")
    ax2.legend()
    ax2.grid(True, alpha=0.3)
    
    # Error bars
    ax3 = fig.add_subplot(133)
    errors = (pred_prices.flatten() - tp_prices.flatten()) * 10000
    bars = ax3.bar(range(len(errors)), errors, color='teal', alpha=0.7)
    for i, b in enumerate(bars):
        if errors[i] < 0: b.set_color('salmon')
    ax3.axhline(0, color='black', linewidth=1)
    ax3.set_title(f"Price Error | MAE: {np.mean(np.abs(errors)):.2f} bps")
    ax3.set_xlabel("Option Index")
    ax3.set_ylabel("Error (bps)")
    ax3.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.savefig('validation_sample.png', dpi=120)
    plt.show()
    
    print(f"\nPrice Error: MAE={np.mean(np.abs(errors)):.2f} bps")
    
    # --- 10. SAVE ARTIFACTS ---
    model.save('heston_calibrator_efficient.keras')
    np.savez('scalers.npz', X_m=X_m, X_s=X_s, Y_m=Y_m, Y_s=Y_s)
    
    print(f"\n{'='*60}")
    print(f"✓ Model saved: heston_calibrator_efficient.keras")
    print(f"✓ Scalers saved: scalers.npz")
    print(f"✓ Cache preserved: {CACHE_DIR}/")
    print(f"{'='*60}\n")
    
    # Memory cleanup
    del X_data, Y_data, P, T_test
    gc.collect()
```
